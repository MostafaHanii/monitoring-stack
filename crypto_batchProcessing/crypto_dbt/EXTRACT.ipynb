```python
def top_coins(limit= '20', key= 'edc9ea32d97240519a256afeb022c3a2'):
    
    url = 'https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest'
    parameters = {
    'start':'1',
    'limit':limit,
    'convert':'USDT'
    }
    headers = {
    'Accepts': 'application/json',
    'X-CMC_PRO_API_KEY': key,
    }

    session = Session()
    session.headers.update(headers)

    try:
        response = session.get(url, params=parameters)
        data = json.loads(response.text)
    except (ConnectionError, Timeout, TooManyRedirects) as e:
        print(e)

    # Normalizes JSON data to dataframe
    df = pd.json_normalize(data['data']) 

    # Adds timestamp column to API data pull
    df['timestamp'] = pd.to_datetime('now') 

    return df

def get_exchangeInfo():
    # filter and get top ones paired with USDT:
    exchange_info = requests.get("https://api.binance.com/api/v3/exchangeInfo").json()
    symbols_df = pd.DataFrame(exchange_info["symbols"])
    # symbols_df[(symbols_df['quoteAsset'] == 'USDT') & (symbols_df['status'] == 'TRADING')][['symbol','baseAsset']]

    symbols_df = symbols_df[(symbols_df['quoteAsset'] == 'USDT') & (symbols_df['status'] == 'TRADING')][['baseAsset']]

    available_coins = list(symbols_df['baseAsset'])
    
    return available_coins

def filter_toBinance_Save(df,binance_coins,file_path= 'DATA/CoinMarketCapTop10Crypto.csv'):
    
    stable_coins = {'USDT', 'USDC', 'BUSD', 'DAI', 'TUSD', 'USTC'}
    top_coins = list(df['symbol'])
    
    binance_top_coins = []
    count = 1 
    
    for coin in top_coins:
        if count > 12:
            break

        if coin in binance_coins:
            if coin not in stable_coins:
                binance_top_coins.append(coin)
                count += 1
        
        if coin == 'USDT':
            binance_top_coins.append(coin)


    print("top coins count: ",len(binance_top_coins))
    
    # Filter original DataFrame to keep only these coins
    df_filtered = df[df['symbol'].isin(binance_top_coins)].copy()

    
    ## Saving Dataframe as CSV File
    
    # Get the current directory
    current_directory = os.getcwd()

    # Construct the file path for saving the CSV file in the current directory
    file_path = os.path.join(current_directory, file_path)

    # Save the DataFrame to a CSV file
    df_filtered.to_csv(file_path, index=False)

    print(f'Data saved to {file_path}')
    return binance_top_coins

top_coins_df = top_coins('30', 'edc9ea32d97240519a256afeb022c3a2')
binance_coins = get_exchangeInfo()
binance_top_coins = filter_toBinance_Save(top_coins_df,binance_coins,file_path= 'DATA/CoinMarketCapTop11Crypto.csv')
binance_top_coins
```

```python
def get_ohlcv(symbol, start_time, end_time, interval='1h'):
    """
    Fetch OHLCV data from Binance for a given symbol and time range.
    """
    url = "https://api.binance.com/api/v3/klines"
    params = {
        "symbol": symbol,
        "interval": interval,
        "startTime": start_time,
        "endTime": end_time,
        "limit": 1000
    }
    response = requests.get(url, params=params)
    if response.status_code != 200:
        print(f" Failed to fetch {symbol}: {response.text}")
        return pd.DataFrame()
    
    data = response.json()
    if not data:
        return pd.DataFrame()

    df = pd.DataFrame(data, columns=[
        "open_time","open","high","low","close","volume",
        "close_time","quote_asset_volume","num_trades",
        "taker_buy_base_volume","taker_buy_quote_volume","ignore"
    ])
    # Convert numeric columns
    for col in ["open","high","low","close","volume"]:
        df[col] = df[col].astype(float)
    # Convert timestamps to datetime
    df["open_time"] = pd.to_datetime(df["open_time"], unit='ms')
    df["close_time"] = pd.to_datetime(df["close_time"], unit='ms')
    return df
```
